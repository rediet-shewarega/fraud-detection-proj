```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

credit_df = pd.read_csv("../data/raw/creditcard.csv")

credit_df.head()
```

```python
credit_df.info()
credit_df.describe()
credit_df.isnull().sum()
credit_df.duplicated().sum()
```

```python
credit_df = credit_df.drop_duplicates()
```

```python
credit_df["Class"].value_counts()
credit_df["Class"].value_counts(normalize=True) * 100
```

```python
sns.countplot(data=credit_df, x="Class")
plt.title("Class Distribution in Credit Card Dataset")
plt.xlabel("Class: 0 = Legitimate, 1 = Fraud")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(credit_df["Amount"], bins=50, kde=True)
plt.title("Distribution of Transaction Amount")
plt.show()
```

```python
sns.boxplot(data=credit_df, x="Class", y="Amount")
plt.title("Transaction Amount by Class")
plt.ylim(0, credit_df["Amount"].quantile(0.95))
plt.show()
```

```python
sns.histplot(credit_df["Time"], bins=50)
plt.title("Distribution of Transaction Time")
plt.show()
```

```python
credit_df.to_csv("../data/processed/creditcard_cleaned.csv", index=False)
```
